```python
1.Create 2 data frames Student and Dept, insert 10 rows to both the table.
Student table has column name: Std_id, name,sem,age, CGPA, dept_id
Department table has columns: Dept_id, Dname, Dloc,D_Hod
```

```python
import pandas as pd
data1={'Std_id':['St11','St12','St13','St14','St15','St16','St17','St18','St19','St20'],
       'name':['Rama','Krishna','Balarama','Hanuma','Jaya','Sita','Radha','Rukmini','Lakshmi',
               'Bharatha'],'Sem':[1,1,2,2,3,3,3,2,1,1],'age':[19,20,21,18,20,22,19,17,19,22],
       'CGPA':[7.5,8.9,9.9,9.5,8.3,8.5,8.3,8.1,9.2,9.5],'Dept_id':['D1','D1','D2','D2','D3',
                                                                   'D4','D4','D4','D1','D2']}
data2={'Dept_id':['D1','D2','D4','D5','D6'],'Dname':['CSE','Mech','AI and ML','ECE','Civil'],
       'Dloc':['Bangalore','Goa','Delhi','Bangalore','Mathura'],'D_Hod':['Indra','Chadra',
                                                                         'Surya','Mourya','Harsha']}
Dept=pd.DataFrame(data2)
Student=pd.DataFrame(data1)
print(Student)
print()
print(Dept)
```

```python
2. Join student and dept tables by using the join method and show the results of both inner and outer join.
```

```python
print(Student.join(Dept.set_index('Dept_id'),on='Dept_id',how='inner'))
print()
print(Student.join(Dept.set_index('Dept_id'),on='Dept_id',how='outer'))
```

```python
#3. Merge student and dept table based on dept_id
```

```python
pd.merge(Student,Dept,on='Dept_id')
```

```python
#4.Extract only CSE department student details.
```
